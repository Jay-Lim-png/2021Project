2. 기하평균  
- 여러개의 수를 연속으로 곱해 그 개수의 거듭제곱근으로 구한 수 
- 기하평균은 인구변동률, 물가상승률과 같이 변화하는 비율을 자료로 제시될 때 사용. 비율의 의미가 있는 %로 제시된 자료에 활용   
- $(X_{1} \times X_{2} \times \ldots \times X_{n})^{1/n}$
- 기하평균 계산 예시
  1. 증가율 공식이 아래와 같을 경우
    - 동아리 회원이 10명으로 시작
    - 1월 인원이 15명일 때 증가율 $r_{1}$ = 15/10 = 1.5
    - 2월 인원이 20명일 때 증가율 $r_{2}$ = 20/15 = 1.3
    - 3월 인원이 25명일 때 증가율 $r_{3}$ = 25/20 = 1.25
  2. 평균증가율 구하는 법  
    1.  최초 인원과 마지막 측정된 인원으로 구하기  (n = 측정된 횟수)  
    r = $\sqrt[n]{\frac{x_{3}}{x_{0}}} = \sqrt[3]{\frac{25}{10}} = 1.36 $  
    2. 증가율만 이용해서 기하평균 공식으로 계산하기  
    r = $\sqrt[3]{r_{1} \times r_{2} \times r_{3}} = \sqrt[3]{1.5 \times 1.3 \times 1.25} = 1.345 $
    3. 산술 평균과의 비교   
    산술평균 $r_{!}, r_{2}, r{3}$ 의 평균을 구하면 1.35로 기하평균과 거의 비슷. 기하평균을 굳이 사용하는 이유는 비율값에 마이너스가 나오는 경우 특히 의미있음. 
  3. 증가율에 0 또는 마이너스가 제시된 경우  
    - 2월 증가율 1.5
    - 3월 증가율 1.33
    - 4월 증가율 -0.25
    1. 증가율이 0 또는 마이너스 일 때 기하평균 공식으로 계산하기  
      마이너스 측정값을 포함해 각 측정값에 1을 더하여 양수값으로 교정하고, 다시 1을 빼줌  
      $r = \sqrt [3] {(1.5 + 1) \times (1.33 + 1) \times (-0.25 + 1)}$ - 1 = 1.63 - 1 = 0.63  
    2. 산술평균과의 비교  
    (1.5 + 1.33 - 0.25) / 3 = 0.86 
  4. 수익률의 기하평균 계산  
    수익률을 계산할 때 다음과 같이 증가율 계산 가능  
    $수익률 = \frac {이후금액 - 이전금액} {이전금액} $  
       - 원래 자본금 100만원
       - 2월 금액 500만원 일 때 증가율 $r_{1} = \frac {500-100} {100}$ = 4 (400%)
       - 3월 금액 200만원일 때 증가율 $r_{2} = \frac {200 - 500} {500}$ = -0.6 (-60%)
       - 4월 금액 200만원일 때 증가율 $r_{3} = \frac {200-200} {200}$ = 0
       - 기하평균 공식으로 수익률 평균 구하기  
        $r = \sqrt[3]{(4+1) \times (-0.6 + 1) \times (0 + 1)} - 1 = 0.25 $(수익률 25%)
     - 산술평균을 이용해 계산한 수익률은 113%로 판단되는데, 결과적으로 투자상품이 100만원에서 200만원이 되었기 때문에 100% 넘는 수익률을 보인다. 그러나 실제적으로는 이 투자과정이 불안정하다는 것을 금액 변화를 통해 알 수 있다. 따라서 산술평균으로 구한 수익률은 불안정한 금액 변화 추이를 제대로 반영하고 있지 못함을 알 수 있다. 

```python
import numpy as np

## 증가율이 음수일 때의 계산 (r1 = 1.5, r2 = 1.33, r3 = -0.25)
### 제곱근을 구할 때는 np.sqrt, 세제곱근을 구할 때는 np.cbrt
np.cbrt([(1.5+1) * (1.33+1) * (-0.25+1)]) - 1
```

3. 조화평균
- 측정값의 역수를 취하여 그 역수를 합한 값으로 평균을 구하는 방식
- 주로 분모를 시간으로 가지는 측정값을 대상으로 함 (전체거리를 다른 속도로 달리는 경우에 한정, 시간의 절반씩 달리면 산술평균 적용)
- 조화평균은 산술/기하 평균에 비해 극단값에 휘둘림이 적고, 평균을 좀 더 잘 찾아내준다
- 조화평균 계산 예시   
전체 거리의 절반을 5km/h의 속도로 달리고, 남은 절반을 10km/h로 달린다면 평균속도는?
- $H = \frac{n}{Σ\frac{1}{x_{i}}} = \frac{2}{\frac{1}{5} + \frac{1} {10}} = 6.66 km/h $
